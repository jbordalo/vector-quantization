# Vector Quantization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage.io import imread
from sklearn.metrics import pairwise_distances_argmin_min
```

## Importing the image

```python
img = imread('https://live.staticflickr.com/65535/48191960402_ac6d65eae8_b.jpg')
```

```python
def show_img(img, title):
    plt.figure(figsize=(12, 8))
    plt.title(title)
    plt.axis('off')
    plt.imshow(img)
```

```python
show_img(img, "Rose (Original)")
```

## Reshaping the data

```python
w, h, d = img.shape
image_array = np.reshape(img / 255.0, (w * h, d))
```

## Quantize the data

```python
n_colors = 8

k_means = KMeans(n_clusters=n_colors).fit(image_array)

centroids = k_means.cluster_centers_
```

## Assign points to closest centroid

```python
indices, _ = pairwise_distances_argmin_min(image_array, centroids)
result = centroids[indices]
```

## Reshape the data

```python
result = np.reshape(result, (w, h, d))
```

## Final result

```python
show_img(result, f"Rose ({n_colors} colors)")
```
